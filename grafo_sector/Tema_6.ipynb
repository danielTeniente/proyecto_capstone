Consiguiendo las posiciones

```python
data_point = pd.read_csv('centros_googleApi')
```

```python
centros = list(data_point['cluster_point'])
```

```python
pos_graph = []
lab_dict = {}
for i,centro in enumerate(centros):
    lab_dict[i] = str(i)
    pos = centro.replace('(','').replace(')','').split(',')
    pos_graph.append((float(pos[0]),float(pos[1])))
```

```python
plt.figure(figsize=(15,15))

nx.draw(time_graph,pos_graph,
            node_size=500,
            labels=lab_dict, 
            with_labels = True)
```

```python
plt.figure(figsize=(15,15))

nx.draw(G,pos_graph,
            node_size=500,
            labels=lab_dict, 
            with_labels = True)
```
