# Tema 1: Uso de Sustractor de fondo

La técnica utilizada para el conteo de vehículos es la extracción de fondo para luego aplicar una detección de objetos por tamaño. 

## OpenCV

Open CV es la biblioteca principal para trabajar con imágenes. Se importa mediante el nombre cv2.

```python
#open CV
import cv2
import matplotlib.pyplot as plt
```

OpenCV permite eliminar el fondo de una imagen o video mediante la función:

```python
backSub = cv2.createBackgroundSubtractorMOG2(history=50, detectShadows=False)
```

Este objeto será usado para eliminar el fondo.
* history toma en cuenta cuántos frames anteiores se usan para considerar lo que es el fondo. Si un objeto aparece en la misma posición durante los mismos frames que el history, el objeto es tomado como fondo.
* detectShadow incluye las sombras como parte del objeto, en este caso es falso, las sombras serán eliminadas como parte del fondo.

## Ejemplo

Se utiliza una imagen de ejemplo de un vehículo visto de frente.

```python
img = cv2.imread('imagenes/vehiculo.jpg')
plt.imshow(img)
```

La imagen no sale con los colores correctos pues OpenCV importa la imagen con orden BGR y pyplot utiliza RGB.

```python
img = cv2.cvtColor(img,cv2.cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

Para elminiar el fondo se utilizar la función apply del objeto backsub.

```python
new_img = backSub.apply(img)
plt.imshow(new_img)
```

En este caso no fue capaz de detectar al objeto pues no tiene frames anteriores de referencia. 

En un video, utiliza frames anteriores para devolver un resultado como el siguiente:

```python
img = cv2.imread('imagenes/autos.PNG',cv2.IMREAD_GRAYSCALE)
plt.imshow(img,'gray')
```

El fondo se ha eliminado y sólo se encuentra la silueta de los vehículos.
