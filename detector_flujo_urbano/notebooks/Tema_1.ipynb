## Puntos ROI

Se utilizan cuatro puntos para delimitar la zona de interés: `x1, x2, y1, y2`. Hay que tomar en cuenta que OpenCV coloca el 0 de `y` en la parte superior.

```python
xr1 = 0
xr2 = 0
yr1 = 0
yr2 = 0
```

Si tenemos una imagen como la siguiente:

```python
img = cv2.imread('imagenes/detectar-taxis.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

Podemos concentrarnos únicamente en la zona izquierda donde los vehículos pasan sobre el puente.

```python
xr1 = 0
xr2 = 1000
yr1 = 500
yr2 = 1200
```

* `x1` se mantiene en 0 porque queremos empezar en la parte izquierda.
* `x2` avanza hasta cerca de la mitad de la imagen
* `y1` empieza en la mitad de la imagen para evitar visualizar el parque que está arriba.
* `y2` llega hasta la parte más baja de la fotografía.

```python
roi = img[yr1:yr2,xr1:xr2]
```

Podemos seleccionar la zona de interés como si se tratara de una matriz en numpy sin problemas.

```python
plt.imshow(roi)
```

Puede verse que se realizó un acercamiento hacia una zona más específica del video.
